Functions

```python
def get_tables(url, text):
    res = requests.get(url)
    comm = re.compile("<!--|-->")
    soup = BeautifulSoup(comm.sub("", res.text), "lxml")
    all_tables = soup.findAll("tbody")
    team_table = all_tables[0]
    team_vs_table = all_tables[1]
    player_table = all_tables[2]
    if text == "for":
        return player_table, team_table
    if text == "vs":
        return player_table, team_vs_table   


def get_frame(features, player_table):
    pre_df_player = dict()
    features_wanted_player = features
    rows_player = player_table.find_all("tr")
    for row in rows_player:
        mins = row.find('td', {'data-stat': 'minutes_90s'})
        if mins != None:
            if float(mins.text.strip().encode().decode("utf-8")) >= 5:
                #if row.find("td", {"data-stat": 'position'}).text.strip().encode().decode("utf-8") != 'GK':
                for f in features_wanted_player:
                    cell = row.find("td", {"data-stat": f})
                    a = cell.text.strip().encode()
                    text = a.decode("utf-8")
                    if text == "":
                        text = "0"
                    if (
                        (f != "player")
                        & (f != "nationality")
                        & (f != "position")
                        & (f != "team")
                        & (f != "age")
                        & (f != "birth_year")
                        & (f != "matches")
                    ):
                        text = float(text.replace(",", ""))
                    if f == 'position':
                        text = 'https://fbref.com' + row.find_all("a")[0]['href']
                    if f in pre_df_player:
                        pre_df_player[f].append(text)
                    else:
                        pre_df_player[f] = [text]
    df_player = pd.DataFrame.from_dict(pre_df_player)
    return df_player


def get_frame_team(features, team_table):
    pre_df_squad = dict()
    features_wanted_squad = features
    rows_squad = team_table.find_all("tr")
    for row in rows_squad:
        if row.find("th", {"data-stat": "team"}) != None:
            name = (
                row.find("th", {"data-stat": "team"})
                .text.strip()
                .encode()
                .decode("utf-8")
            )
            if "team" in pre_df_squad:
                pre_df_squad["team"].append(name)
            else:
                pre_df_squad["team"] = [name]
            for f in features_wanted_squad:
                cell = row.find("td", {"data-stat": f})
                if cell != None:
                    a = cell.text.strip().encode()
                    text = a.decode("utf-8")
                    if text == "":
                        text = "0"
                    if (
                        (f != "player")
                        & (f != "nationality")
                        & (f != "position")
                        & (f != "team")
                        & (f != "age")
                        & (f != "birth_year")
                        & (f != "matches")
                    ):
                        text = float(text.replace(",", ""))
                    if f in pre_df_squad:
                        pre_df_squad[f].append(text)
                    else:
                        pre_df_squad[f] = [text]
    df_squad = pd.DataFrame.from_dict(pre_df_squad)
    return df_squad


def frame_for_category(category, top, end, features):
    url = top + category + end
    player_table, team_table = get_tables(url, "for")
    df_player = get_frame(features, player_table)
    return df_player


def frame_for_category_team(category, top, end, features, text):
    url = top + category + end
    player_table, team_table = get_tables(url, text)
    df_team = get_frame_team(features, team_table)
    return df_team


def get_outfield_data(top, end):
    df1 = frame_for_category("stats", top, end, stats_player)
    df2 = frame_for_category("shooting", top, end, shooting_player)
    df3 = frame_for_category("passing", top, end, passing_player)
    df4 = frame_for_category("passing_types", top, end, passing_types_player)
    df5 = frame_for_category("gca", top, end, gca_player)
    df6 = frame_for_category("defense", top, end, defense_player)
    df7 = frame_for_category("possession", top, end, possession_player)
    df8 = frame_for_category("misc", top, end, misc_player)
    df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8], axis=1)
    df = df.loc[:, ~df.columns.duplicated()]
    return df


def get_keeper_data(top, end):
    df1 = frame_for_category("keepers", top, end, keepers_player)
    df2 = frame_for_category("keepersadv", top, end, keepersadv_player)
    df = pd.concat([df1, df2], axis=1)
    df = df.loc[:, ~df.columns.duplicated()]
    return df


def get_team_data(top, end, text):
    df1 = frame_for_category_team("stats", top, end, stats_team, text)
    df2 = frame_for_category_team("keepers", top, end, keepers_team, text)
    df3 = frame_for_category_team("keepersadv", top, end, keepersadv_team, text)
    df4 = frame_for_category_team("shooting", top, end, shooting_team, text)
    df5 = frame_for_category_team("passing", top, end, passing_team, text)
    df6 = frame_for_category_team("passing_types", top, end, passing_types_team, text)
    df7 = frame_for_category_team("gca", top, end, gca_team, text)
    df8 = frame_for_category_team("defense", top, end, defense_team, text)
    df9 = frame_for_category_team("possession", top, end, possession_team, text)
    df10 = frame_for_category_team("misc", top, end, misc_team, text)
    df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10], axis=1)
    df = df.loc[:, ~df.columns.duplicated()]
    return df
```
